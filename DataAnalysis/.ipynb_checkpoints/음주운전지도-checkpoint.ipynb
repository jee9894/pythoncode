```python
import folium
import pandas as pd
import requests
import json
import webbrowser
```

```python
filename = '시도구군별_음주운전_데이터_2005_2019.csv'
df = pd.read_csv(filename, header=[0], index_col=[0,1], skipfooter=1, encoding='utf-8')
df
```

```python
df1 = df.xs('합계',level=1)
df1
```

```python
df1 = df1.rename(index={'서울':'서울특별시','부산':'부산광역시','대구':'대구광역시',
                      '광주':'광주광역시','대전':'대전광역시','인천':'인천광역시',
                      '울산':'울산광역시','세종':'세종특별자치시','경기':'경기도',
                      '강원':'강원도','충북':'충청북도','충남':'충청남도','전남':'전라남도',
                      '경북':'경상북도','경남':'경상남도','제주':'제주특별자치도','전북':'전라북도'})
```

```python
df1['2005']
```

```python
year = '2019'
```

```python
map_osm = folium.Map(location=[37.566345,126.977893],zoom_start=7)
rfile = open('skorea-provinces-2018-geo.json', 'r', encoding='utf-8').read()
jsonData = json.loads(rfile)
folium.Choropleth(geo_data=jsonData,
                  data=df1[year],
                  columns=[df1.index,year],
                  key_on='feature.properties.name',
                  fill_color='YlGn',
                  fill_opacity=0.7,
                  line_opacity=0.3,
                  legend_name='음주사고건수').add_to(map_osm)
map_osm.save('test.html')
```
