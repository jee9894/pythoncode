```python
import pandas as pd
```

```python
df1 = pd.read_excel('시도,시군구별_음주정도별_교통사고_2015_2019.xls', sheet_name='Sheet1', header=[0,1,2], index_col=[0,1,2])
df1.head(10)
```

```python
df1 = df1.xs('합계', level=1, axis=1).xs('사고건수', level=2)
df1.head(10)
```

```python
df2 = pd.read_excel('시도,시군구별_음주정도별_교통사고_2010_2014.xls', sheet_name='Sheet1', header=[0,1,2], index_col=[0,1,2])
```

```python
df2 = df2.xs('합계', level=1, axis=1).xs('사고건수', level=2)
df2.head(10)
```

```python
df3 = pd.read_excel('시도,시군구별_음주정도별_교통사고_2005_2009.xls', sheet_name='Sheet1', header=[0,1,2], index_col=[0,1,2])
df3 = df3.xs('합계', level=1, axis=1).xs('사고건수', level=2)
df3.head(10)
```

```python
df = pd.concat([df3,df2,df1], axis=1)
df.head(10)
```

```python
df
```

```python
df = df.fillna(0)
```

```python
df.loc['세종']
```

```python
df = df.replace({'2005': '-', '2006' : '-', '2007' : '-', '2008' : '-', '2009' : '-', '2010' : '-', '2011' : '-', '2012' : '-', '2013' : '-', '2014' : '-', '2015' : '-', '2016' : '-', '2017' : '-', '2018' : '-', '2019' : '-'}, 
                {'2005': 0, '2006' : 0, '2007' : 0, '2008' : 0, '2009' : 0, '2010' : 0, '2011' : 0, '2012' : 0, '2013' : 0, '2014' : 0, '2015' : 0, '2016' : 0, '2017' : 0, '2018' : 0, '2019' : 0})
```

```python
df = df.astype('int')
```

```python
df = df.xs('합계', level=1, axis=1)
```

```python
df.columns.name = ''
```
