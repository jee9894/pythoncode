```python
import requests

r = requests.get('https://finance.naver.com//marketindex/exchangeList.nhn')
soup = BeautifulSoup(r.text,'html.parser')
soup
```

```python
table = soup.select('.tbl_exchange')
table
```

```python
tit = soup.select('.tit')
```

```python
sales = soup.select('.sale')
```

```python
list_tit = []
for i in tit:
    list_tit.append(i.string.strip())

list_sales = []
for i in sales:
    list_sales.append(i.string.replace(",",""))
```

```python
result = list(zip(list_tit, list_sales))
result
```

```python
r = requests.get('https://news.naver.com/')
soup = BeautifulSoup(r.text, 'html.parser')
soup
```

```python
aside = soup.select('.main_aside .section_wide')
aside
```

```python
categories = soup.select('#category_ranking > a')
categories
```

```python
top10s = soup.select('.section_list_ranking > li > a')
top10s[59].string
```

```python
list = []
for news in categories:
    for i in range(0,10):
        list.append((news.string, top10s[0].string))
        del top10s[0]
list
```

```python
results = soup.select('.section_list_ranking > li > a')
for result in results:
    print('제목 :', result.attrs['title'])
    print('링크 :', result.attrs['href'])
    print()
    url_content = 'https://news.naver.com/'+result.attrs['href']
    response_content = requests.get(url_content)
    soup_content = BeautifulSoup(response_content.text, 'html.parser')
    content = soup_content.select_one('#articleBodyContents')
#     print(content.contents)
    
    output = ''
    for item in content.contents:
        stripped = str(item).strip()
        if stripped == '':
            continue
        if stripped[0] not in ['<', '/']:
            output += str(item).strip()
    print(output.replace('본문 내용TV플레이어',''))
#     print(output)
    print()
    break
```
