```python
def net_get_geo_coder(query):
    category_group_code = 'PO3'
    KAKAO_API_KEY = '51cc9fa14a42b26c2569cc0a0050de45'
    url = '''
    https://dapi.kakao.com/v2/local/search/keyword.json?query={}&category_group_code={}
    '''.format(query, category_group_code)
    headers = {'Authorization' : 'KakaoAK {0}'.format(KAKAO_API_KEY)}
    res = requests.get(url, headers=headers)
    try:
        return (float(res.json()['documents'][0]['y']), float(res.json()['documents'][0]['x']))
    except:
        return False
```

```python
result = net_get_geo_coder('제주특별자치도 제주시 중앙로7길 15 (일도일동)')
result
```

```python
map_osm = folium.Map(location=[33.39004277172384,126.52573908075342], zoom_start=11)
for index, row in df.iterrows():
    print(index)
    geoData = net_get_geo_coder(row['주소'])
    print(geoData)
    if geoData != None and geoData is not False:
        folium.Marker(geoData, popup=row['읍면동'].strip(), icon=folium.Icon(color='red')).add_to(map_osm)
map_osm
```

```python
svFilename = 'jeju.html'
map_osm.save(svFilename)

import webbrowser
webbrowser.open(svFilename)
```
